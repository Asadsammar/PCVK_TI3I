```python
import numpy as np
import cv2 as cv
import pandas as pd
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
from skimage import transform
import matplotlib.pyplot as plt
from PIL import Image
```

```python
url= "https://placekitten.com/800/571"
img = io.imread(url)
```

```python
print('Change the brightness level of the image')
print('----------------------------------------')
try:
  brightness = int(input('Typr Brightness Value'))
except ValueError:
  print('Error, Not a valid Numner')

brightness_image = np.zeros(img.shape, img.dtype)

#access pixel by pixel
for y in range(img.shape[0]):
  for x in range(img.shape[1]):
    for c in range(img.shape[2]):
      brightness_image[y,x,c] = np.clip(img[y,x,c] + brightness, 0, 255)


image01 =cv.resize(img, (0,0),fx=0.5,fy=0.5)
image02 =cv.resize(brightness_image, (0,0),fx=0.5,fy=0.5)
finalFram = cv.hconcat((image01, image02))
cv2_imshow(finalFram)

# finaFram = cv.hconcat((img, brightness_image))
# cv2_imshow(finaFram)
```

```python
print('Implement image inverse')
print('-----------------------')
inverted_image = 255 - img

image1 =cv.resize(img, (0,0),fx=0.5,fy=0.5)
image2 =cv.resize(inverted_image, (0,0),fx=0.5,fy=0.5)

finaFram = cv.hconcat((image1, image2))
cv2_imshow(finaFram)
```

```python

print('Adjust Brightness and Contrast of the Image')
print('-------------------------------------------')

try:
    brightness = int(input('Type Brightness Value: '))
    contrast = float(input('Type Contrast Value: '))
except ValueError:
    print('Error: Invalid input')



if img is None:
    print('Error: Unable to load image')
else:
    # Adjust brightness
    brightness_image = np.zeros(img.shape, img.dtype)
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            for c in range(img.shape[2]):
                brightness_image[y, x, c] = np.clip(img[y, x, c] + brightness, 0, 255)

    # Adjust contrast
    contrast_image = np.zeros(img.shape, img.dtype)
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            for c in range(img.shape[2]):
                contrast_image[y, x, c] = np.clip((img[y, x, c] - 127.5) * contrast + 127.5, 0, 255)

    # Resize original and adjusted images
    image01 = cv.resize(img, (0, 0), fx=0.5, fy=0.5)
    image02 = cv.resize(brightness_image, (0, 0), fx=0.5, fy=0.5)
    image03 = cv.resize(contrast_image, (0, 0), fx=0.5, fy=0.5)

    # Concatenate the images side by side
    feeFrame = cv.hconcat((image02, image03))

    # Display the final image
    cv2_imshow(image01)
    cv2_imshow(feeFrame)
```

```python
# Apply Grayscale Effects
grayscale_average = np.mean(img, axis=2).astype(np.uint8)
grayscale_lightness = ((np.max(img, axis=2) + np.min(img, axis=2)) / 2).astype(np.uint8)
grayscale_luminance = ((0.21 * img[:, :, 0]) + (0.72 * img[:, :, 1]) + (0.07 * img[:, :, 2])).astype(np.uint8)


# Resize original and adjusted images
img04 = cv.resize(grayscale_average, (0, 0), fx=0.5, fy=0.5)
img05 = cv.resize(grayscale_lightness, (0, 0), fx=0.5, fy=0.5)
image06 = cv.resize(grayscale_luminance, (0, 0), fx=0.5, fy=0.5)

# Concatenate the images side by side
fFrame = cv.hconcat((img04, img05))

# Display the final image
cv2_imshow(img04)
cv2_imshow(img05)
cv2_imshow(image06)



```
