**Show images in Red-Green and Green-Blue channels only!**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

# Load the image
img=cv.imread('/content/drive/MyDrive/PVCK_images/kitten01.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

# Create new images with only Red-Green and Green-Blue channels
red_green_image = img.copy()
red_green_image[:, :, 2] = 0  # Set the Blue channel to zero

green_blue_image = img.copy()
green_blue_image[:, :, 0] = 0  # Set the Red channel to zero

# Create subplots to display the original and modified images
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# Original image
axes[0].imshow(img)
axes[0].set_title('Original')

# Red-Green channel image
axes[1].imshow(red_green_image)
axes[1].set_title('Red-Green Channel Only')

# Green-Blue channel image
axes[2].imshow(green_blue_image)
axes[2].set_title('Green-Blue Channel Only')

plt.tight_layout()
plt.show()
```

**Show image rows 5-30, all columns, channel Green only!**

```python
img=cv.imread('/content/drive/MyDrive/PVCK_images/kitten01.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

# Define the region of interest (ROI) rows 5-30, all columns, and the green channel
roi = img[5:31, :, 1]  # Channel 1 represents the green channel (0-indexed)

# Display the green channel of the ROI
plt.imshow(roi, cmap='gray')  # Use 'gray' colormap to display as grayscale
plt.title('Green Channel (Rows 5-30)')
plt.show()
```

Create 5 boxes of different sizes and colors in one image. it is recommended to use

```python
img=cv.imread('/content/drive/MyDrive/PVCK_images/kitten01.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

cv.rectangle(img, pt1=(100,10),pt2=(10,100),color=(0,255,0),thickness=10)
cv.rectangle(img, pt1=(100,400),pt2=(10,310),color=(255,0,0),thickness=5)
cv.rectangle(img, pt1=(100,250),pt2=(10,150),color=(255,0,255),thickness=8)
cv.rectangle(img, pt1=(200,350),pt2=(10,150),color=(0,255,255),thickness=5)
cv.rectangle(img, pt1=(100,450),pt2=(10,350),color=(255,255,255),thickness=7)
plt.imshow(img)
```
