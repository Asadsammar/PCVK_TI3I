```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('/content/drive/MyDrive/PVCK_images/lena.jpg')

# Split the image into its color channels
channels = cv2.split(image)

colors = ("b", "g", "r")
channel_names = ("Blue", "Green", "Red")

# Create bins for histogram
bins = np.arange(256)

# Create a single vertical figure for all histograms
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# Loop over each of the color channels
for i, (channel, color, name) in enumerate(zip(channels, colors, channel_names)):
    # Create a histogram for the current channel and plot it
    histogram = cv2.calcHist([channel], [0], None, [256], [0, 256])

    ax[i].set_title(f"{name} Histogram for lena.jpg")
    ax[i].set_xlabel("Bins")
    ax[i].set_ylabel("# of Pixels")
    ax[i].bar(bins, histogram.flatten(), color=color, edgecolor='none')

plt.tight_layout()
plt.show()
```

7. Do the mapping process of 16 million RGB colors into just 8 colors, namely black,
green, yellow, blue, cyan, magenta, white as shown in the following output based on
the flowchart below! (Use the image testlena .jpg)

```python
import numpy as np
import cv2

# Load image
img = cv2.imread('/content/drive/MyDrive/PVCK_images/lena.jpg')

# Define 8 colors
colors = np.array([[0,0,0], [0,255,0], [255,255,0], [0,0,255], [255,255,255],
                   [255,0,255], [0,255,255], [255,255,255]])

# Get image shape
height, width, channels = img.shape

# Create color index map
index_map = np.zeros((height,width), np.uint8)

# Map each pixel to closest color
for y in range(height):
  for x in range(width):
    B, G, R = img[y,x]
    min_dist = 1000000
    for i in range(len(colors)):
      dist = abs(R - colors[i,0]) + abs(G - colors[i,1]) + abs(B - colors[i,2])
      if dist < min_dist:
        min_dist = dist
        index_map[y,x] = i

# Create output image
output = colors[index_map]

# Display input and output image
plt.imshow(img)
plt.show()

plt.imshow(output)
plt.show()
```

8. Do the Floyd and Steinberg dithering process like the following output (initial image
display, and after dithering display) based on the flowchart below! (Use wiki.jpg
image)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/PVCK_images/wiki.jpg', 0)

height, width = img.shape

for y in range(height-1):
  for x in range(1, width-1):

    oldPixel = img[y,x]

    # Threshold to 1 bit (0 or 255)
    newPixel = 0 if oldPixel < 128 else 255

    img[y,x] = newPixel

    quantError = oldPixel - newPixel

    # Floyd-Steinberg diffusion
    img[y, x+1] = min(255, img[y, x+1] + quantError*7/16)
    img[y+1, x-1] = min(255, img[y+1, x-1] + quantError*3/16)
    img[y+1, x] = min(255, img[y+1, x] + quantError*5/16)
    img[y+1, x+1] = min(255, img[y+1, x+1] + quantError*1/16)

# Display input and output image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
plt.title('Original')
plt.show()

plt.imshow(img, cmap='gray')
plt.title('Dithered')
plt.show()
```
