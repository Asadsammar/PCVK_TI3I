```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('/content/drive/MyDrive/PVCK_images/lena.jpg')

# Split the image into its color channels
channels = cv2.split(image)

colors = ("b", "g", "r")
channel_names = ("Blue", "Green", "Red")

# Create bins for histogram
bins = np.arange(256)

# Create a single vertical figure for all histograms
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# Loop over each of the color channels
for i, (channel, color, name) in enumerate(zip(channels, colors, channel_names)):
    # Create a histogram for the current channel and plot it
    histogram = cv2.calcHist([channel], [0], None, [256], [0, 256])

    ax[i].set_title(f"{name} Histogram for lena.jpg")
    ax[i].set_xlabel("Bins")
    ax[i].set_ylabel("# of Pixels")
    ax[i].bar(bins, histogram.flatten(), color=color, edgecolor='none')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('/content/drive/MyDrive/PVCK_images/lena.jpg')

# Split the image into its color channels
channels = cv2.split(image)

colors = ("b", "g", "r")
channel_names = ("Blue", "Green", "Red")

# Create bins for histogram
bins = np.arange(256)

# Create a single vertical figure for all histograms
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# Loop over each of the color channels
for i, (channel, color, name) in enumerate(zip(channels, colors, channel_names)):
    # Calculate histogram using NumPy
    histogram, _ = np.histogram(channel, bins=256, range=(0, 256))

    ax[i].set_title(f"{name} Histogram for lena.jpg")
    ax[i].set_xlabel("Bins")
    ax[i].set_ylabel("# of Pixels")
    ax[i].bar(bins, histogram, color=color, edgecolor='none')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the RGB image
image = cv2.imread('/content/drive/MyDrive/PVCK_images/lena.jpg')

# Split the image into its color channels
blue, green, red = cv2.split(image)

# Perform histogram equalization for each channel
blue_equalized = cv2.equalizeHist(blue)
green_equalized = cv2.equalizeHist(green)
red_equalized = cv2.equalizeHist(red)

# Create bins for histogram
bins = np.arange(256)

# Create separate vertical figures for each channel's histograms
fig, axs = plt.subplots(2, 3, figsize=(15, 8))

# Plot histograms for the original image
axs[0, 0].hist(blue.ravel(), bins=256, color='b', alpha=0.5)
axs[0, 0].set_title('Blue Channel (Original)')
axs[0, 1].hist(green.ravel(), bins=256, color='g', alpha=0.5)
axs[0, 1].set_title('Green Channel (Original)')
axs[0, 2].hist(red.ravel(), bins=256, color='r', alpha=0.5)
axs[0, 2].set_title('Red Channel (Original)')

# Plot histograms for the equalized image
axs[1, 0].hist(blue_equalized.ravel(), bins=256, color='b', alpha=0.5)
axs[1, 0].set_title('Blue Channel (Equalized)')
axs[1, 1].hist(green_equalized.ravel(), bins=256, color='g', alpha=0.5)
axs[1, 1].set_title('Green Channel (Equalized)')
axs[1, 2].hist(red_equalized.ravel(), bins=256, color='r', alpha=0.5)
axs[1, 2].set_title('Red Channel (Equalized)')

plt.tight_layout()
plt.show()
```
